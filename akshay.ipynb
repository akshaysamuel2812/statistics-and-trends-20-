```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Automobile.csv')  # Update path as needed
origin_counts = data['origin'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(origin_counts, labels=origin_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Cars by Origin')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['weight'], data['horsepower'], alpha=0.7, color='Red', edgecolor='')
plt.xlabel('Weight')
plt.ylabel('Horsepower')
plt.title('Horsepower vs. Weight')
plt.show()
```
